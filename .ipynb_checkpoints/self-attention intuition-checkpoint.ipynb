# What does *self* refer to in self-attention?

"Self-attention" mechanism, like other mechanisms introduces some weights. However, the key difference is how the weights are defined. Before getting into that, let's get the understanding about what **"self"** refers to in the term **self-attention**. To do so, we go through a simple example.

Let's assume that we want to assess a car company's performance by looking at the price of the cars it makes. We look at the prices over a 10-year period.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
year=range(10)
price=[20,22,23,26,25,26,30,28,30,27]
price_history = pd.DataFrame(zip(year,price), 
                             columns=['year','price(k)'])
ax=sns.lineplot(x='year', y='price(k)', data=price_history, marker="*", markersize=10)
_=ax.axvline([7], color='red')
```

The trend is noisy; while it is upward overall, for example at year 7 the price dropped. If we want to look at the trend globally this noise is better to be removed. In other words, if I want to only look at the year 7, which shows a price drop, and still be able to make a good judgment about the overall  performance of the company, I should get rid of the noise.
We may do that by smoothing it by applying moving average or any low pass filter. Let's take the following low pass filter, which is also called exponential moving average:

$$\hat{y}_{t}= (1-\alpha) \times \hat{y}_{t-1} + \alpha\times y_t $$

so at each year $t$, a new representation for the price $y_t$ is the weighted mean of the prices in preceding years.

Let's take a look at the output:


```python
price_history['price_smoothed(k)'] = price_history['price(k)'].ewm(alpha=.5).mean()

ax=sns.lineplot(x='year', y='price(k)', data=price_history, marker="*", markersize=10, label='Original data')
sns.lineplot(ax=ax, x='year', y='price_smoothed(k)', data=price_history, marker="o", markersize=10, label='Smoothed representation')
```

Now, what we did above could also be named attention: for each observation in the 10 year sample, we calculate a new representation which is based on weighing on other observations. Through $\alpha$ we decide how much to *attend* to the previous values and the current value.

## Contextualization

In the example above, we contextualize each observation in the 10-year sample. The context, however, is only time. In a more general scenario, each observation could be associated with more context information, which could also be named as attributes. 

Assume that the car company only makes two car models A and B, and the cars are only in two colors black and blue. Let's add this information to the price history:

```python
price_history['color']=['black', 'black', 'black','blue', 'black','blue','black','blue','black','blue']
price_history['model']=['A','A','A','A','A','A','B','B','A','B']
```

Now we have more context to consider when looking at the company performance, let's incorporate the models and colors :

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.lineplot(ax=ax[0], x='year', y='price(k)', data=price_history, hue='model',marker="*", markersize=10)
sns.lineplot(ax=ax[1], x='year', y='price(k)', data=price_history, hue='color',marker="*", markersize=10)
```
